```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conn = sqlite3.connect("loterias.db")
```

```python
megasena = pd.read_sql("SELECT * FROM megasena", conn)
megasena.head()
```

```python
megasena.columns
```

```python
megasena.describe()
```

```python
megasena.info()
```

```python
colunas_dezenas = [c for c in megasena.columns if c.startswith("dezena_")]

frequencias = megasena[colunas_dezenas].stack().value_counts().sort_index()

plt.figure(figsize=(18,5))
sns.barplot(x=frequencias.index, y=frequencias.values)
plt.title("Frequência das Dezenas - Mega-Sena")
plt.xlabel("Dezena")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(megasena["qtd_pares"], kde=False, bins=10)
plt.title("Distribuição de Quantidade de Pares por Concurso")
plt.show()

plt.figure(figsize=(8,4))
sns.histplot(megasena["qtd_impares"], kde=False, bins=10)
plt.title("Distribuição de Quantidade de Ímpares por Concurso")
plt.show()
```
